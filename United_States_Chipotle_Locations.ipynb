Importing Data

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jeffreybraun/chipotle-locations")

print("Path to dataset files:", path)
```

```python
import os
dataset_files = os.listdir(path)
print("Dataset files:", dataset_files)
```

```python
import pandas as pd

csv_file_path = os.path.join(path, 'chipotle_stores.csv')
df = pd.read_csv(csv_file_path)

print(df.head())  #getting a feel of the data head
```

```python
import json
import os

#define the path to the US states GeoJSON file
us_states_file = os.path.join(path, 'us-states.json')

#open and load the JSON file!
with open(us_states_file) as f:
    states_geojson = json.load(f)

#inspect the first few elements of the GeoJSON structure to make sure it looks good to code with
print(states_geojson)
```

Checking for any missing values

```python
#first lets check for any missing values or issues in the latitude and longitude columns just incase
print(df[['latitude', 'longitude']].isnull().sum())  #result is 0 so there is none
```

First checking to see if it will plot the states lines from the json

```python
import folium

#lets do a Folium map centered around the US - (googled)
map_center = [37.0902, -95.7129]
chipotle_map = folium.Map(location=map_center, zoom_start=5) #centering

#add the US states boundaries from the loaded GeoJSON data just to make sure they look good before plotting loctions
folium.GeoJson(states_geojson).add_to(chipotle_map)

#displaying map
from IPython.display import display
display(chipotle_map)
```

```python
import pandas as pd
import folium
import json
from IPython.display import display

#another map centered around the US
map_center = [37.0902, -95.7129]
chipotle_map = folium.Map(location=map_center, zoom_start=5)

#adding the US states boundaries to the map like before
folium.GeoJson(states_geojson).add_to(chipotle_map)

#add CircleMarker for each Chipotle store location per the longitude and latitude coordinate
for index, row in df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']

    #check if the coordinates are valid just incase
    if pd.notnull(latitude) and pd.notnull(longitude):
        folium.CircleMarker(
            location=[latitude, longitude],
            radius=1,  #want a small size of the dot
            color='red',  #making color to red
            fill=True,  #filled
            fill_color='blue',  # border color
            fill_opacity=0.8,  #opacity of the fill ... almost fully
        ).add_to(chipotle_map)

#displaying the map !!
display(chipotle_map)
```
